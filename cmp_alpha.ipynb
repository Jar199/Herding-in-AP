```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def normalize(data):
    return (data - np.min(data)) / (np.max(data) - np.min(data))
```

### compare empircal alpha value to theoretical one for monthly pattern

```python
emp = pd.read_csv('extreme_alpha_monthly_90.csv')
theo = pd.read_csv('theo_monthly.csv')
emp_a = normalize(emp['alpha'][1:])
th_a = normalize(theo['alpha'])
th_c = normalize(theo['alpha_c'])
plt.scatter(emp_a, th_a, marker='o', color='blue')
plt.scatter(emp_a, th_c, marker='o', color='orange')
plt.show()
```

공분산을 계산하여 상관계수 계산. 

```python
print("상관계수 between empirical alpha and theoretical alpha 'without c'", np.cov(emp_a, th_a)[0, 1])
print("상관계수 between empirical alpha and theoretical alpha 'with c'", np.cov(emp_a, th_c)[0, 1])
```
