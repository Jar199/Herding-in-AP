```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

analysis monthly pattern for extreme prices  
: c, alpha, alpha with c

```python
df = pd.read_csv('theo_monthly.csv')
df['date'] = df['date'].astype('str')
date = []
for i in range(len(df)):
    date.append(df['date'][i][4:])
params = ['c', 'alpha', 'alpha_c']
plt.rcParams['font.family'] = 'Times New Roman'
for p in params:
    s, e = 0, 12
    for year in range(2016, 2023):
        plt.figure(figsize=(14, 10))
        plt.title(year, size=45, pad=15)
        plt.plot(date[s:e], df[p][s:e], marker='o', linewidth=7, markersize=8, label=p)
        plt.xlabel('date(month)', fontsize=35, labelpad=10)
        plt.ylabel(p, fontsize=35, labelpad=10)
        plt.xticks(fontsize=25)
        #plt.yticks([np.min(df[p]), np.max(df[p])], fontsize=25)
        plt.yticks(fontsize=25)
        plt.legend(fontsize=20)
        #plt.show()
        plt.savefig(f"./fig/fig_theo_month/{p}_{year}")
        s = e
        e += 12
```

analysis yearly pattern for extreme prices  
: c, alpha, alpha with c

```python
df = pd.read_csv('theo_yearly_final_b.csv')
df['date'] = df['date'].astype('str')

params = ['c', 'alpha', 'alpha_c']
plt.rcParams['font.family'] = 'Times New Roman'
for p in params:
    plt.figure(figsize=(14, 10))
    plt.title('2016-2022', size=45, pad=15)
    plt.plot(df['date'], df[p], marker='o', linewidth=7, markersize=8, label=p)
    plt.xlabel('date(year)', fontsize=35, labelpad=10)
    plt.ylabel(p, fontsize=35, labelpad=10)
    plt.xticks(fontsize=25)
    plt.yticks(fontsize=25)
    plt.legend(fontsize=20)
    #plt.show()
    plt.savefig(f"./fig/fig_theo_year/{p}")
```
